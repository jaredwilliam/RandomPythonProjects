# Collatz Conjecture

```python
import matplotlib.pyplot as plt
```

```python
def collatz(x):
    nums = []
    nums.append(x)
    while x != 1:
        if x % 2 == 1:
            x = x*3+1
            nums.append(int(x))
        else:
            x = x / 2
            nums.append(int(x))
    return nums
```

```python
def plot_collatz(number):
    f, ax = plt.subplots(figsize = (16,8))
    for num in range(1,number):
        ax.plot(collatz(num))
    ax.set_xlabel('Num Steps to Get to 1')
    ax.set_ylabel('Value')
    return ax
```

```python
plot_collatz(500)
```
